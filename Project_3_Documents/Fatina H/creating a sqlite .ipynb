```python
import sqlite3
import pandas as pd

# Path to the CSV file
data_path = 'Resources/global_youtube_statistics.csv'

# Read the CSV file into a DataFrame
try:
    data = pd.read_csv(data_path, encoding='utf-8')
except UnicodeDecodeError:
    try:
        data = pd.read_csv(data_path, encoding='latin1')
    except UnicodeDecodeError:
        data = pd.read_csv(data_path, encoding='ISO-8859-1')

# Display the first few rows of the DataFrame
data.head()

# Create a SQLite database connection
conn = sqlite3.connect('youtube_data.db')
cursor = conn.cursor()

# Create table
cursor.execute('''
CREATE TABLE IF NOT EXISTS youtube_data (
    rank INTEGER,
    Youtuber TEXT,
    subscribers INTEGER,
    video_views INTEGER,
    category TEXT,
    Title TEXT,
    uploads INTEGER,
    Country TEXT,
    Abbreviation TEXT,
    channel_type TEXT,
    video_views_rank INTEGER,
    country_rank INTEGER,
    channel_type_rank INTEGER,
    video_views_for_the_last_30_days INTEGER,
    lowest_monthly_earnings INTEGER,
    highest_monthly_earnings INTEGER,
    lowest_yearly_earnings INTEGER,
    highest_yearly_earnings INTEGER,
    subscribers_for_last_30_days INTEGER,
    created_year INTEGER,
    created_month INTEGER,
    created_date DATE,
    Gross_tertiary_education_enrollment REAL,
    Population INTEGER,
    Unemployment_rate REAL,
    Urban_population REAL,
    Latitude REAL,
    Longitude REAL
)
''')

# Insert data into the table
data.to_sql('youtube_data', conn, if_exists='replace', index=False)

# Commit and close the connection
conn.commit()
conn.close()
```
